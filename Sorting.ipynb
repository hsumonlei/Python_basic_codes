**Bubble Sort**

```python
# Bubble Sort is the simplest sorting algorithm that works by 
# repeatedly swapping the adjacent elements if they are in the wrong order.


str_arr = input("Enter your input array num1, num2, num3... :").split(', ') #will take in a string of numbers separated by a space
print(str_arr)
arr = [int(num) for num in str_arr]
print(arr)

#********Bubble Sort***********
for j in range(len(arr)):
    print("\nOuter Loop :", j)
    for i in range(1,len(arr)):
         print(arr[i] <= arr[i-1])
         if (arr[i] <= arr[i-1]) :
              temp = arr[i]
              arr[i] = arr[i-1]
              arr[i-1] = temp
         print(arr)


print("\nSorted array : \n",arr)
for i in range(len(arr)):
    print("%d" % arr[i], end=" ")
          
```

**Worst and Average Case Time Complexity: O(n*n) or O(n^2). The worst case occurs when an array is reverse sorted.**

**Best Case Time Complexity: O(n). The best case occurs when an array is already sorted.**
    
Space Complexity: O(1)
Boundary Cases: Bubble sort takes minimum time (Order of n) when elements are already sorted.
Sorting In Place: Yes
Stable: Yes
Due to its simplicity, bubble sort is often used to introduce the concept of a sorting algorithm. 

**Selection Sort**

The selection sort algorithm sorts an array by repeatedly finding the minimum element (considering ascending order) from unsorted part and putting it at the beginning. The algorithm maintains two subarrays in a given array.
1) The subarray which is already sorted. 
2) Remaining subarray which is unsorted.
In every iteration of selection sort, the minimum element (considering ascending order) from the unsorted subarray is picked and moved to the sorted subarray. 

```python
str_arr = input("Enter your input array num1, num2, num3... :").split(', ') #will take in a string of numbers separated by a space
print(str_arr)
arr = [int(num) for num in str_arr]
print(arr)

#********Selection Sort***********

for i in range(0,len(arr)) :
    min_index = i
    print("Minimum Array index:",i)
    for j in range(i+1, len(arr)):
        if arr[min_index] > arr[j]:
            min_index = j
    temp = arr[i]
    arr[i] = arr[min_index]
    arr[min_index] = temp
    print(arr)


print("\nSorted array : \n",arr)
for i in range(len(arr)):
    print("%d" % arr[i], end=" ")
```

**Worst , Best and Average Case Time Complexity: O(n*n) or O(n^2).**
Space Complexity: O(1)

Note :The good thing about selection sort is it never makes more than O(n) swaps and can be useful when memory write is a costly operation.
